**Stacking - İstifleme**

```python
import numpy as np
```

```python
dizi1 = np.array([[1,2],[3,4]])
dizi1
```

```python
dizi2 = np.array([[-1,-2],[-3,-4]])
dizi2
```

```python
dizi_dikey = np.vstack((dizi1,dizi2))
dizi_dikey
```

```python
dizi_yatay = np.hstack((dizi1,dizi2))
dizi_yatay
```

```python
#problem 2 boyutlu 2 satır ve 2 sütün olan 3 adet array oluştur,
#bunları yatayda ve dikeyde birleştir.
```

```python
dizi3 = np.array([[1,3],[4,6]])
dizi3
```

```python
dizi_dikey_uclu = np.vstack((dizi1,dizi2,dizi3))
dizi_dikey_uclu
```

```python
dizi_yatay_uclu = np.hstack((dizi1,dizi2,dizi3))
dizi_yatay_uclu
```
